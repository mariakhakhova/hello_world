```python
import numpy as np
import matplotlib.pyplot as plt

table = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt", dtype ={'names': ('year', 'hare', 'lynx', 'carrot'), 
                                  'formats': ('i8', 'f8', 'f8', 'f8')})

```

```python
plt.plot(table['year'], table['carrot'], color = 'orange')
plt.plot(table['year'], table['lynx'], color = 'red')
plt.plot(table['year'], table['hare'], color = 'grey')
plt.legend(('carrot', 'lynx', 'hare'))
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
data_histogram = table['hare'] + table['lynx'] 
data_histogram = table['lynx'] / data_histogram * 100
plt.bar (table['year'], data_histogram, color='red')
plt.title('lynx')
plt.xlabel('Year')
plt.ylabel('pct')
plt.show()

data_histogram = 100 - data_histogram
plt.bar (table['year'], data_histogram, color='black')
plt.ylabel('pct')
plt.xlabel('Year')
plt.title('hare')
plt.show()
```

```python
print('Mean lynx: {}'.format(np.mean(table['lynx'])))
print('Std lynx: {}'.format(np.std(table['lynx'])))
print('Mean lynx: {}'.format(np.mean(table['carrot'])))
print('Std lynx: {}'.format(np.std(table['carrot'])))
print('Mean lynx: {}'.format(np.mean(table['hare'])))
print('Std lynx: {}'.format(np.std(table['hare'])))

plt.hist(table['lynx'], 10, color = (1, 0, 0, 1))
plt.axvline(x=np.mean(table['lynx']), color = 'black')
plt.title('lynx')
plt.show()

plt.hist(table['carrot'], 10, color= 'orange')
plt.axvline(x=np.mean(table['carrot']), color = 'blue')
plt.title('carrot')
plt.show()

plt.hist(table['hare'], 10, color = 'grey')
plt.axvline(x=np.mean(table['hare']), color = 'red')
plt.title('hare')
plt.show()
```

```python
print('argmax hare: {}'.format(table['year'][table['hare'].argmax()]))
print('argmax carrot: {}'.format(table['year'][table['carrot'].argmax()]))
print('argmax lynx: {}'.format(table['year'][table['lynx'].argmax()]))

plt.plot(table['year'], table['hare'], 'grey')
plt.plot(table['year'], table['carrot'], 'orange')
plt.plot(table['year'], table['lynx'], 'red')
plt.axvline(x = table['year'][table['hare'].argmax()], color='grey', linestyle = ':')
plt.axvline(x = table['year'][table['carrot'].argmax()], color='orange', linestyle = ':')
plt.axvline(x = table['year'][table['lynx'].argmax()], color='red', linestyle = ':')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'carrot', 'lynx'))
plt.show()
```

```python
a = np.sign( - table['lynx'] + table['hare'])
b = list(map(lambda x: 'H' if x == 1 else 'L', a))
print(b)
plt.pie([b.count('H'),b.count('L')], labels=['H', 'L'], autopct='%1.1f%%', shadow = 1)
plt.show()
```

```python
print('Above N for hare: {}'.format(table['year'][np.where(table['hare'] > 50000)]))
print('Above N for carrot: {}'.format(table['year'][np.where(table['carrot'] > 50000)]))
print('Above N for lynx: {}'.format(table['year'][np.where(table['lynx'] > 50000)]))

plt.plot(table['year'], table['carrot'], 'orange')
plt.plot(table['year'], table['hare'], 'grey')
plt.plot(table['year'], table['lynx'], 'red')

plt.vlines(table['year'][np.where(table['carrot'] > 50000)], 0, 100000, color = 'orange', linestyle = ':')
plt.vlines(table['year'][np.where(table['hare'] > 50000)], 0, 100000, color = 'grey', linestyle = ':')
plt.vlines(table['year'][np.where(table['lynx'] > 50000)], 0, 100000, color = 'red', linestyle = ':')

plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('carrot', 'hare', 'lynx'))
plt.axis([1900, 1920, 0, 100000])
plt.show()
```

```python
print('For hare: {}'.format(table['year'][np.argsort(table['hare'])[::-1][19:]]))
print('For carrot: {}'.format(table['year'][np.argsort(table['carrot'])[::-1][19:]]))
print('For lynx: {}'.format(table['year'][np.argsort(table['lynx'])[::-1][19:]]))
```

```python
var = table['hare'][1:] - table['hare'][:20]

print('Correlation =', np.corrcoef(var, table['lynx'][1:])[0][1])

plt.plot(table['year'][1:], table['lynx'][1:], 'red')
plt.plot(table['year'][1:], var , 'black')

plt.legend(('lynx', 'var',))
plt.axis([1901, 1920, -100000, 100000])
plt.show()

```
